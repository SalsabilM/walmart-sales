```python
#QC
# Define holiday dates
holidays = {
    'Super_Bowl': ['12/2/2010', '11/2/2011', '10/2/2012'],
    'Labour_Day': ['10/9/2010', '9/9/2011', '7/9/2012'],
    'Thanksgiving': ['26-11-2010', '25-11-2011', '23-11-2012'],
    'Christmas': ['31-12-2010', '30-12-2011', '28-12-2012']
}

max_sales = 0
max_sales_aid = ""

for holiday, dates in holidays.items():
    holiday_sales = df[df['Date'].isin(dates)]['Weekly_Sales']
    max_holiday_sales = holiday_sales.max()
    if max_holiday_sales > max_sales:
        max_sales = max_holiday_sales
        max_sales_aid = holiday

print("The aid with the highest sales during holidays is:", max_sales_aid)
```

```python
def remove_outliers(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return column[(column >= lower_bound) & (column <= upper_bound)]
    
columns_to_filter = ['Weekly_Sales', 'Temperature', 'Unemployment']

for col in columns_to_filter:
    df[col] = remove_outliers(df[col])

# Columns to plot boxplots for
columns_to_plot = ['Store', 'Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']

# Create subplots for boxplots
fig, axes = plt.subplots(nrows=1, ncols=len(columns_to_plot), figsize=(16, 6))

# Plot boxplots
for i, column in enumerate(columns_to_plot):
    df.boxplot(column=column, ax=axes[i])
    axes[i].set_title(f'Boxplot for {column} Column')
    axes[i].set_ylabel('Values')

plt.tight_layout()
plt.show()
```
